```python
import pandas as pd
import os
import glob

# Set the folder path where your CSVs are located
folder_path = "/Users/chandhiny/Revised Code CS7CS5/Dataset"  # Change this!

# Find all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Load and concatenate all CSVs
dataframes = []
for file in csv_files:
    df = pd.read_csv(file)
    df["source_file"] = os.path.basename(file)  # Optional: track source file
    dataframes.append(df)

# Merge all into one DataFrame
merged_df = pd.concat(dataframes, ignore_index=True)

# Save the merged data
merged_df.to_csv("merged_output.csv", index=False)

print("Merged", len(csv_files), "CSV files into 'merged_output.csv'")
```

```python
import pandas as pd

merged_df = pd.read_csv("/Users/chandhiny/Revised Code CS7CS5/merged_output.csv", thousands=',')

merged_df['ArchiveDate'] = pd.to_datetime(merged_df['ArchiveDate'], errors='coerce')

merged_df = merged_df.dropna(subset=['ArchiveDate'])

monthly_df = merged_df.groupby('ArchiveDate')['Total'].sum().reset_index()

monthly_df['Month Year'] = monthly_df['ArchiveDate'].dt.strftime('%m-%Y')

monthly_df = monthly_df.rename(columns={'Total': 'Total Patients Waiting List'})

final_df = monthly_df[['Month Year', 'Total Patients Waiting List']]

final_df.to_csv("/Users/chandhiny/Revised Code CS7CS5/MONTHLY_INPATIENT_NATIONAL_TOTAL.csv", index=False)
```
